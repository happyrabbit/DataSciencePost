```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1, x2 = np.meshgrid(np.linspace(-13.2, 13.2, 50),
                     np.linspace(-37.4, 37.4, 50))
```

```python
v = (1/2) * x1 + (np.sqrt(3) / 2) * x2
g1 = -1 / (1 + np.exp(-v))
```

```python
fig = plt.figure(figsize = (12, 12))
ax = fig.add_subplot(2, 2, 1, projection = '3d')
surf = ax.plot_surface(x1, x2, g1)
```

```python
v = x1
g2 = (v + 5) * np.sin(1 / (v / 3 + 0.1))
```

```python
ax = fig.add_subplot(2, 2, 2, projection = '3d')
surf = ax.plot_surface(x1, x2, g2)
```

```python
v = x2
g3 = (np.exp(v ** 2/5))
```

```python
ax = fig.add_subplot(2, 2, 3, projection = '3d')
surf = ax.plot_surface(x1, x2, g3)
```

```python
v = x1
g4 = (v + 0.1) * np.sin(1 / (v / 3 + 0.1))
```

```python
ax = fig.add_subplot(2, 2, 4, projection = '3d')
surf = ax.plot_surface(x1, x2, g4)
plt.show()
```
