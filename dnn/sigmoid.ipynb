```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v = np.linspace(-10, 10, 200)
s1 = 1
s2 = 0.3
s3 = 20
h1 = 1 / (1 + np.exp(-s1 * v))
h2 = 1 / (1 + np.exp(-s2 * v))
h3 = 1 / (1 + np.exp(-s3 * v))
```

```python
fig, ax = plt.subplots()
line1 = ax.plot(v, h1, label = 's = 1')
line2 = ax.plot(v, h2, label = 's = 0.3')
line3 = ax.plot(v, h3, label = 's = 20')
ax.legend(loc = 'upper left')
plt.show()
```
